# 15. File & Document Processing Tools

These tools let an agent **read your files** — text files, PDFs, CSVs — and answer questions about them.

## Common file tools (from crewai-tools)

| Tool | Reads |
|------|-------|
| `FileReadTool` | A plain text file (.txt) |
| `PDFSearchTool` | A PDF file |
| `CSVSearchTool` | A CSV (spreadsheet) |
| `DOCXSearchTool` | A Word document |

Install once: `pip install crewai-tools`

## Step 1: Create a small text file to read

```python
text = """Our shop is open from 9 AM to 6 PM.
We sell shoes, bags, and belts.
Returns are allowed within 7 days."""

with open("shop_info.txt", "w") as f:
    f.write(text)

print("shop_info.txt created!")
```

## Step 2: Give a file-reading tool to an agent

```python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, Task, Crew, LLM

from crewai_tools import FileReadTool

llm = LLM(model="gpt-4o-mini")

# Point the tool at our file
file_tool = FileReadTool(file_path="shop_info.txt")

assistant = Agent(
    role="Shop Assistant",
    goal="Answer questions using the shop file",
    backstory="You read the shop file and answer customers.",
    tools=[file_tool],
    llm=llm,
)
print("Assistant ready!")
```

## Step 3: Ask a question about the file

```python
task = Task(
    description="Using the file, what time does the shop close?",
    expected_output="The closing time.",
    agent=assistant,
)

crew = Crew(agents=[assistant], tasks=[task])

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
```

## Key points to remember

- File tools let an agent **read your files** and answer about them.
- `FileReadTool` reads text; `PDFSearchTool`, `CSVSearchTool`, `DOCXSearchTool` read other types.
- You can point a tool at a file: `FileReadTool(file_path="shop_info.txt")`.
- Give it to an agent with `tools=[file_tool]`.
- This is the first step toward letting agents work with **your own documents**.
