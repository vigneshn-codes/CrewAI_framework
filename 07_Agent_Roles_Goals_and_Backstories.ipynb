# 7. Agent Roles, Goals & Backstories

These three lines **shape how an agent behaves**. Good ones = better answers.

- **Role** = the job title.
- **Goal** = what success looks like.
- **Backstory** = the personality and experience.

## Why they matter

The AI reads these words and **acts the part**. A clear role gives focused answers; a vague one gives messy answers.

### Real-life analogy
Telling a chef *"You are a healthy-food chef who avoids sugar"* gives very different meals than just *"cook something."*

## Weak vs strong examples

| Part | Weak (vague) | Strong (clear) |
|------|--------------|----------------|
| role | "Helper" | "Senior Python Teacher" |
| goal | "Help people" | "Explain Python to total beginners simply" |
| backstory | "You know stuff" | "You taught 1000 beginners and love clear examples" |

## See the difference in code

Here are two agents. Notice how the **strong** one will give better, more focused replies.

```python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, LLM

llm = LLM(model="gpt-4o-mini")

# Strong, clear agent
teacher = Agent(
    role="Senior Python Teacher",
    goal="Explain Python to total beginners in very simple words",
    backstory="You have taught 1000 beginners and love clear, friendly examples.",
    llm=llm,
)

print("Role:", teacher.role)
print("Goal:", teacher.goal)
```

## Simple tips for writing good ones

- **Role:** use a real job title ("Market Researcher", not "AI").
- **Goal:** start with a verb ("Find...", "Write...", "Explain...").
- **Backstory:** add experience + style ("You are careful and always use simple words").
- Keep them **short and clear** — one sentence each is fine.

## Key points to remember

- **Role, goal, backstory** control how an agent behaves.
- **Clear** wording gives much better results than vague wording.
- Role = job title, Goal = what success looks like, Backstory = personality + experience.
- Start goals with a verb and keep everything short and simple.
