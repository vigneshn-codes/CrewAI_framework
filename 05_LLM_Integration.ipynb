# 5. LLM Integration (OpenAI, Gemini, Claude, Ollama)

The **LLM** is the **brain** of an agent — the AI that thinks and writes.

CrewAI can use many brains. In this course we mainly use **OpenAI**, but switching to another is just **one line**.

## The easy way: just set OPENAI_API_KEY

If your `OPENAI_API_KEY` is set (in your `.env`), CrewAI uses OpenAI automatically — you don't have to write anything extra.

But it's good to be **clear** and pick the model yourself, like below.

## Main brain: OpenAI

````python
from dotenv import load_dotenv
load_dotenv()

from crewai import LLM

# Our main AI brain for the whole course
llm = LLM(model="gpt-4o-mini")

print("OpenAI brain is ready!")
````

## Using this brain in an agent

You pass the brain to an agent with `llm=...`.

````python
from crewai import Agent

helper = Agent(
    role="Helper",
    goal="Answer questions simply",
    backstory="You explain things to beginners.",
    llm=llm,          # <-- the brain we made above
)

print("Agent created with the OpenAI brain!")
````

## Switching to other brains (only the model line changes)

| Provider | Code | Needs |
|----------|------|-------|
| **OpenAI** | `LLM(model="gpt-4o-mini")` | `OPENAI_API_KEY` |
| **Google Gemini** | `LLM(model="gemini/gemini-1.5-flash")` | `GEMINI_API_KEY` |
| **Claude** | `LLM(model="claude-3-5-sonnet-20240620")` | `ANTHROPIC_API_KEY` |
| **Ollama (local, free)** | `LLM(model="ollama/llama3", base_url="http://localhost:11434")` | Ollama app installed |

### Examples (run only the one you have a key for)
```python
from crewai import LLM

# Google Gemini  (set GEMINI_API_KEY)
gemini_llm = LLM(model="gemini/gemini-1.5-flash")

# Claude  (set ANTHROPIC_API_KEY)
claude_llm = LLM(model="claude-3-5-sonnet-20240620")

# Ollama - runs on your own computer, no key needed
ollama_llm = LLM(model="ollama/llama3", base_url="http://localhost:11434")
```

Notice: only the **model name** changes. The rest of your CrewAI code stays the same.

## Key points to remember

- The **LLM** is the agent's **brain**.
- `from crewai import LLM` then `LLM(model="gpt-4o-mini")` for OpenAI.
- Pass it to an agent with `llm=...`.
- To use Gemini, Claude, or Ollama, **only the model name changes**.
- **Ollama** runs locally and needs **no API key** (great for free practice).
