# 6. Understanding Agents

An **Agent** is one AI worker. It has a **role** (who it is), a **goal** (what it wants), and a **brain** (the LLM).

This is the most important building block in CrewAI.

## Real-life analogy

An agent is like a **new employee** on their first day. You tell them:
- **Who they are** ("You are our cook.")
- **What to achieve** ("Make tasty meals.")
- **Their background** ("You trained in Italy.")

With that, they know how to behave.

## The 3 must-have parts of an agent

| Part | Question it answers | Example |
|------|---------------------|---------|
| **role** | Who are you? | "Travel Guide" |
| **goal** | What is your aim? | "Suggest fun places" |
| **backstory** | What's your background? | "You have toured 50 countries" |

## Step 1: Set up the brain

```python
from dotenv import load_dotenv
load_dotenv()

from crewai import LLM

llm = LLM(model="gpt-4o-mini")
print("Brain ready!")
```

## Step 2: Create an agent

```python
from crewai import Agent

guide = Agent(
    role="Travel Guide",
    goal="Suggest fun places to visit",
    backstory="You have toured 50 countries and love helping travelers.",
    llm=llm,
    verbose = True
)

print("Agent created:", guide.role)
```

## A few optional settings

| Setting | What it does | Default |
|---------|--------------|---------|
| `verbose=True` | Print what the agent is thinking | False |
| `allow_delegation=False` | Stop it from passing work to others | False |
| `tools=[...]` | Give it extra powers (later notebook) | none |

An agent alone does nothing yet — it needs a **Task** (next notebook).

## Key points to remember

- An **Agent** is one AI worker.
- It needs 3 things: **role**, **goal**, and **backstory**.
- You give it a brain with `llm=...`.
- `verbose=True` lets you watch it think.
- An agent needs a **Task** before it can actually do anything.
