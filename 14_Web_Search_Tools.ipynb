# 14. Web Search Tools

The AI brain only knows what it learned during training. A **web search tool** lets an agent look up **fresh, up-to-date** information online.

## Why we need it

- The model does not know **today's** news or prices.
- A search tool fetches **current** facts from the internet.

### Real-life analogy
It's like letting an employee **Google something** instead of guessing from memory.

## Step 1: Get a free search key (Serper)

CrewAI's `SerperDevTool` uses **serper.dev** (free tier) for Google search.

1. Sign up at **https://serper.dev**.
2. Copy your API key.
3. Add it to your `.env` file:
```
SERPER_API_KEY=your-serper-key
```
Also install tools: `pip install crewai-tools`

## Step 2: Create the search tool and give it to an agent

````python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, Task, Crew, LLM

from crewai_tools import SerperDevTool

llm = LLM(model="gpt-4o-mini")

# The web search tool
search_tool = SerperDevTool()

researcher = Agent(
    role="Researcher",
    goal="Find current information from the web",
    backstory="You search the internet for fresh facts.",
    tools=[search_tool],
    llm=llm,
)
print("Researcher with web search is ready!")
````

## Step 3: Ask it something that needs fresh info

````python
task = Task(
    description="Search the web and tell me one recent news headline about AI.",
    expected_output="One recent headline with a short note.",
    agent=researcher,
)

crew = Crew(agents=[researcher], tasks=[task])

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
````

## Key points to remember

- A **web search tool** lets an agent find **fresh** info online.
- `SerperDevTool` (from `crewai-tools`) does Google search.
- It needs a free **`SERPER_API_KEY`** in your `.env`.
- Give it to an agent with `tools=[search_tool]`.
- The agent searches automatically when the task needs current facts.
