# 9. Creating Multi-Step Workflows

A **workflow** is just **many tasks done in order**, where each step builds on the one before.

Real jobs are rarely one step — so we give the crew a **list of tasks**.

## Real-life analogy

Making a YouTube video:
```
Step 1: Research the topic
Step 2: Write the script   (uses the research)
Step 3: Make a title       (uses the script)
```
Each step uses the result of the step before. That is a workflow.

## How CrewAI passes results between steps

By default, the output of one task is **automatically given as context** to the next task in the list. So step 2 can use what step 1 produced — you don't wire it up by hand.

## Step 1: One agent, the brain

````python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, Task, Crew, LLM

llm = LLM(model="gpt-4o-mini")

writer = Agent(
    role="Content Writer",
    goal="Create simple, clear content",
    backstory="You write friendly content for beginners.",
    llm=llm,
)
print("Agent ready!")
````

## Step 2: Make a list of tasks (the steps)

````python
# Step A: list ideas
task1 = Task(
    description="List 3 fun facts about the moon.",
    expected_output="3 short fun facts.",
    agent=writer,
)

# Step B: use those facts to write a tweet
task2 = Task(
    description="Use the facts above to write one short tweet.",
    expected_output="One short tweet.",
    agent=writer,
)
print("Two steps ready!")
````

## Step 3: Run the workflow (tasks run in order)

````python
crew = Crew(
    agents=[writer],
    tasks=[task1, task2],   # they run top to bottom
)

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
````

## Key points to remember

- A **workflow** is many tasks done **in order**.
- Put tasks in a **list**: `tasks=[task1, task2, ...]`.
- Each task's output is **automatically passed** to the next task.
- This lets later steps build on earlier results.
- The same agent can do all steps, or different agents can (next notebook).
