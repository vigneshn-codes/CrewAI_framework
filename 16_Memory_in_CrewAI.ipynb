# 16. Memory in CrewAI

**Memory** lets a crew **remember things** during its work, instead of forgetting everything after each step.

With memory on, agents can recall earlier facts and give better, more consistent answers.

## Real-life analogy

Without memory, an agent is like someone with **no notes** — they forget what was said a minute ago.

With memory, they **keep a notebook** and can look back at what happened earlier.

## Turning memory on is easy

Just add **`memory=True`** when you create the Crew. CrewAI handles the rest.

```python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, Task, Crew, LLM

llm = LLM(model="gpt-4o-mini")

helper = Agent(
    role="Helper",
    goal="Help and remember details",
    backstory="You remember what the user told you.",
    llm=llm,
)

task = Task(
    description="Greet the user and remember their name is Thiru.",
    expected_output="A friendly greeting.",
    agent=helper,
)

crew = Crew(
    agents=[helper],
    tasks=[task],
    memory=True,    # <-- turn memory ON
)

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
```

## What memory helps with

- Remembering facts the user gave earlier.
- Staying consistent across many tasks.
- Not repeating the same work twice.

> Note: memory uses embeddings behind the scenes, so it may make an extra API call. That's normal.

## Key points to remember

- **Memory** lets a crew remember things during its work.
- Turn it on with **`memory=True`** in the Crew.
- It helps agents stay consistent and recall earlier facts.
- Without memory, agents forget after each step.
