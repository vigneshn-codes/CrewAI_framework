# 13. Integrating Custom Python Tools

Sometimes no ready-made tool fits your need. So you build your **own** tool — just a normal Python function the agent can call.

## The simple idea

1. Write a normal Python function.
2. Add the **`@tool`** decorator on top.
3. Give it to an agent.

That's it — now the agent can use your function as a power.

## Step 1: Make a custom tool

The text in the `"""..."""` (the docstring) tells the agent **what the tool does**, so write it clearly.

```python
from dotenv import load_dotenv
load_dotenv()

from crewai.tools import tool

@tool("Add Tool")
def add_numbers(a: int, b: int) -> int:
    """Add two numbers together and return the result."""
    return a + b

print("Custom tool created!")
```

## Step 2: Give the tool to an agent

```python
from crewai import Agent, Task, Crew, LLM

llm = LLM(model="gpt-4o-mini")

math_agent = Agent(
    role="Math Helper",
    goal="Solve simple math using tools",
    backstory="You always use your tools to calculate.",
    tools=[add_numbers],     # <-- our custom tool
    llm=llm,
)
print("Agent has the custom tool!")
```

## Step 3: Use it in a task

```python
task = Task(
    description="What is 25 plus 17? Use your tool to add them.",
    expected_output="The number answer.",
    agent=math_agent,
)

crew = Crew(agents=[math_agent], tasks=[task])

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
```

## Key points to remember

- A **custom tool** is just a Python function with the **`@tool`** decorator.
- Import it: `from crewai.tools import tool`.
- The **docstring** explains the tool to the agent — write it clearly.
- Give it to an agent with `tools=[your_function]`.
- The agent calls your function automatically when it helps the task.
