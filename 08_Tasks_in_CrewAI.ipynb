# 8. Tasks in CrewAI

A **Task** is one **job** you give to an agent.

An agent is the worker; a task is the **assignment** that tells the worker exactly what to do.

## The 3 main parts of a task

| Part | Meaning |
|------|---------|
| **description** | What to do (the instruction) |
| **expected_output** | What the finished result should look like |
| **agent** | Which agent will do it |

### Real-life analogy
A task is like a **sticky note** on someone's desk: *"Write a 100-word summary of this report — leave it on my desk by 5pm."*

## Step 1: Make an agent (the worker)

```python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, LLM

llm = LLM(model="gpt-4o-mini")

writer = Agent(
    role="Writer",
    goal="Write short, simple text",
    backstory="You write clearly for beginners.",
    llm=llm,
)
print("Writer ready!")
```

## Step 2: Make a task (the job)

```python
from crewai import Task

write_poem = Task(
    description="Write a short 2-line poem about the sun.",
    expected_output="A simple 2-line poem.",
    agent=writer,
)
print("Task ready!")
```

## Step 3: Run it with a Crew

A task needs a **Crew** to actually run. (We learn Crews in detail soon.)

```python
from crewai import Crew

crew = Crew(agents=[writer], tasks=[write_poem])

# Inside a Jupyter notebook we must use kickoff_async() with "await".
# (A plain .py script uses crew.kickoff() instead.)

result = await crew.kickoff_async()

#result = crew.kickoff() for .py script
print(result)
```

## Key points to remember

- A **Task** is one job given to an agent.
- Its 3 main parts: **description**, **expected_output**, **agent**.
- `expected_output` tells the agent what the finished result should look like.
- A task runs inside a **Crew** using `crew.kickoff()`.
- One agent can be given many tasks.
