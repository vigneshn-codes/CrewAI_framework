# 1. Introduction to AI Agents and Agentic AI

An **AI Agent** is an AI that can **think, decide, and take actions** on its own to reach a goal — not just answer one question.

**Agentic AI** means building apps where one or more agents **work step by step** (and sometimes use tools) to finish a bigger job.

## A normal chatbot vs an AI agent

| Normal chatbot | AI Agent |
|----------------|----------|
| Answers one question | Works toward a goal |
| Waits for you each time | Plans the next step itself |
| Cannot use tools | Can search the web, read files, use tools |
| One reply and done | Many steps until the job is finished |

## Real-life analogy

Think of hiring a **personal assistant**:

- You say: *"Plan my weekend trip."*
- The assistant **figures out the steps**: check weather, find hotels, book a cab.
- They **use tools** (phone, websites) and come back with the finished plan.

An AI agent works the same way — you give a **goal**, and it does the steps.

## What an agent is made of

| Part | Meaning | Example |
|------|---------|---------|
| **Role** | Who the agent is | "Travel Planner" |
| **Goal** | What it must achieve | "Plan a cheap weekend trip" |
| **Brain (LLM)** | The AI that thinks | OpenAI GPT |
| **Tools** | Extra powers | Web search, calculator |
| **Task** | The exact job to do | "Find 3 hotels under $100" |

## Why use multiple agents?

Big jobs are easier when split among a **team**. Just like a company has different people for different work:

- A **Researcher** agent finds information.
- A **Writer** agent turns it into a report.

This "team of agents" idea is exactly what **CrewAI** helps us build (next notebook).

## Key points to remember

- An **AI Agent** can think, decide, and take actions to reach a goal.
- **Agentic AI** = apps where agents work step by step, often using tools.
- A normal chatbot just answers; an agent **works until the job is done**.
- An agent has a **role, goal, brain (LLM), tools, and tasks**.
- Many agents can work together as a **team** — this is what CrewAI is for.
