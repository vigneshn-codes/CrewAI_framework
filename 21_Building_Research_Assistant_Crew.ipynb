# 21. Building a Research Assistant Crew

Now we put it all together! We build a small **2-agent crew** that researches a topic and writes a short report.

- **Researcher** agent → finds the facts.
- **Writer** agent → turns them into a report.

## The plan

```
Researcher  ->  finds 3 key points about the topic
     |
Writer      ->  turns those points into a short report
```

This is a real, useful mini-app — and it uses everything you've learned.

## Step 1: Setup and two agents

````python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, Task, Crew, LLM

llm = LLM(model="gpt-4o-mini")

researcher = Agent(
    role="Researcher",
    goal="Find the most important facts about a topic",
    backstory="You are great at finding clear, true facts.",
    llm=llm,
)

writer = Agent(
    role="Report Writer",
    goal="Write a short, simple report",
    backstory="You turn facts into easy-to-read reports.",
    llm=llm,
)
print("Two agents ready!")
````

## Step 2: Two tasks (research, then write)

````python
topic = "the benefits of drinking water"

research_task = Task(
    description=f"Find 3 key points about {topic}.",
    expected_output="3 short bullet points.",
    agent=researcher,
)

write_task = Task(
    description="Use the 3 points to write a short 5-sentence report.",
    expected_output="A simple 5-sentence report.",
    agent=writer,
)
print("Tasks ready!")
````

## Step 3: Build the crew and run it

````python
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
)

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
````
