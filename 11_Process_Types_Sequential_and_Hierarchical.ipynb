# 11. Process Types (Sequential & Hierarchical)

The **Process** decides **how** the crew works:

- **Sequential** = tasks run one after another (in order).
- **Hierarchical** = a **manager** agent decides who does what.

## The two types

| Process | How it works | Real-life match |
|---------|--------------|-----------------|
| **Sequential** | Task 1 → Task 2 → Task 3, in order | A to-do list |
| **Hierarchical** | A manager assigns and checks work | A boss leading a team |

Sequential is the **simplest** and the default. Beginners should start here.

## Sequential (the easy, default one)

```python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, Task, Crew, Process, LLM

llm = LLM(model="gpt-4o-mini")

writer = Agent(role="Writer", goal="Write simply", backstory="You write for beginners.", llm=llm)

task = Task(description="Write one fun fact about cats.", expected_output="One fun fact.", agent=writer)

crew = Crew(
    agents=[writer],
    tasks=[task],
    process=Process.sequential,   # run tasks in order
)

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
```

## Hierarchical (a manager runs the team)

Here CrewAI creates a **manager** that hands out tasks and reviews the work. You must give it a `manager_llm` (the manager's brain).

```python
researcher = Agent(role="Researcher", goal="Find facts", backstory="You find facts.", llm=llm)
writer2 = Agent(role="Writer", goal="Write text", backstory="You write text.", llm=llm)

job = Task(
    description="Research and then write 3 lines about the ocean.",
    expected_output="3 short lines.",
    # note: no agent set - the manager decides who does it
)

crew = Crew(
    agents=[researcher, writer2],
    tasks=[job],
    process=Process.hierarchical,
    manager_llm=llm,            # the manager's brain (required here)
)

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
```

## Which should I use?

- **Start with Sequential** — it is simple and predictable.
- Use **Hierarchical** only for bigger, complex jobs where a manager should decide the order.

## Key points to remember

- **Process** = how the crew works.
- **Sequential** runs tasks in order (default, easiest).
- **Hierarchical** uses a **manager** agent to assign and review work.
- Hierarchical needs a `manager_llm`.
- Beginners should start with **Sequential**.
