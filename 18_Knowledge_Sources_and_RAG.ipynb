# 18. Knowledge Sources & RAG

A **Knowledge Source** is your **own information** (text, files) that you give to a crew so its agents can answer using **your** data.

This is **RAG** (Retrieval-Augmented Generation): the agent **looks up your knowledge first**, then answers.

## Why use it?

The AI brain does not know about **your** company, notes, or rules. A knowledge source teaches it your facts — without retraining the model.

### Real-life analogy
It's like giving a new employee your **company handbook** so they answer like they work for you.

## Step 1: Create a knowledge source from text

The easiest type is a **StringKnowledgeSource** — just some text you write.

```python
from dotenv import load_dotenv
load_dotenv()

from crewai.knowledge.source.string_knowledge_source import StringKnowledgeSource

facts = """Our company is called BlueShoe.
We were founded in 2019.
We sell eco-friendly shoes made from recycled plastic."""

knowledge = StringKnowledgeSource(content=facts)
print("Knowledge source ready!")
```

## Step 2: Give the knowledge to the Crew

```python
from crewai import Agent, Task, Crew, LLM

llm = LLM(model="gpt-4o-mini")

agent = Agent(
    role="Company Helper",
    goal="Answer using only company knowledge",
    backstory="You know all about BlueShoe.",
    llm=llm,
)

task = Task(
    description="What does our company sell, and when were we founded?",
    expected_output="A short answer from the company facts.",
    agent=agent,
)

crew = Crew(
    agents=[agent],
    tasks=[task],
    knowledge_sources=[knowledge],   # <-- attach your knowledge
)

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
```

## What happened (the RAG idea)

1. The agent **searched your knowledge** for the answer.
2. It found the BlueShoe facts.
3. It answered using **your** information, not a guess.

That is RAG inside CrewAI — and it's mostly automatic.

## Key points to remember

- A **Knowledge Source** is your own data given to the crew.
- This enables **RAG**: look up your facts, then answer.
- `StringKnowledgeSource(content=...)` is the simplest one.
- Attach it with `knowledge_sources=[...]` on the Crew.
- Agents then answer from **your** information, not just their training.
