# 10. Crew and Agent Collaboration

**Collaboration** means **many agents working together**, each doing the part they are best at.

The **Crew** is the team that holds all the agents and tasks.

## Real-life analogy

Making a newspaper:
- A **Reporter** finds the news.
- An **Editor** turns it into a clean article.

Two people, two skills, one finished result. CrewAI works the same way.

## Step 1: Create two different agents

```python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, Task, Crew, LLM

llm = LLM(model="gpt-4o-mini")

reporter = Agent(
    role="Reporter",
    goal="Find key facts about a topic",
    backstory="You gather clear, true facts.",
    llm=llm,
)

editor = Agent(
    role="Editor",
    goal="Turn facts into a short, neat article",
    backstory="You write polished, simple articles.",
    llm=llm,
)
print("Two agents ready!")
```

## Step 2: Give each agent its own task

```python
find_facts = Task(
    description="Find 3 key facts about electric cars.",
    expected_output="3 short facts.",
    agent=reporter,
)

write_article = Task(
    description="Use the facts to write a 4-sentence article.",
    expected_output="A short 4-sentence article.",
    agent=editor,
)
print("Tasks ready!")
```

## Step 3: Put them in one Crew and run

```python
crew = Crew(
    agents=[reporter, editor],
    tasks=[find_facts, write_article],
)

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
```

## What happened?

1. The **Reporter** found the facts.
2. Those facts were **passed** to the **Editor**.
3. The **Editor** wrote the final article.

The agents collaborated to produce one result — just like a real team.

## Key points to remember

- **Collaboration** = many agents, each doing their best part.
- The **Crew** holds all the agents and tasks.
- Each agent gets its **own task**.
- One agent's output is **passed** to the next, so they build on each other.
- This teamwork is the real power of CrewAI.
