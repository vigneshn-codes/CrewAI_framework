# 4. Installing and Setting Up CrewAI

Before we build anything, we install CrewAI and give it an **API key** so the AI brain works.

## Step 1: Install CrewAI

In the terminal:
```
pip install crewai crewai-tools
```

- `crewai` = the main framework.
- `crewai-tools` = ready-made tools (web search, file reading, etc.).

## Step 2: Get an OpenAI API key

1. Go to **https://platform.openai.com**.
2. Create an **API key**.
3. Keep it safe (never share it).

## Step 3: Save the key in a `.env` file

The easiest way: make a file named **`.env`** next to your notebook and put your key inside:
```
OPENAI_API_KEY=your-key-here
```

Then install the reader:
```
pip install python-dotenv
```

## Step 4: Load the key in your code

This reads the `.env` file and makes the key available to CrewAI.

````python
from dotenv import load_dotenv

load_dotenv()   # reads the .env file

print("Setup done! CrewAI can now use your OpenAI key.")
````

## Step 5: Quick test

If this import works with no error, CrewAI is installed correctly.

````python
from crewai import Agent, Task, Crew

print("CrewAI is ready to use!")
````

## Key points to remember

- Install with `pip install crewai crewai-tools`.
- CrewAI needs an **API key** for its AI brain (we use OpenAI).
- Save the key in a **`.env`** file: `OPENAI_API_KEY=your-key-here`.
- Load it in code with `from dotenv import load_dotenv` and `load_dotenv()`.
- Test the setup with `from crewai import Agent, Task, Crew`.
