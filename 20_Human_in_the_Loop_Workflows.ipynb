# 20. Human-in-the-Loop Workflows

**Human-in-the-loop** means a **person checks or approves** the agent's work before it is finished.

This keeps a human in control for important decisions.

## Why we need it

Agents are helpful but not perfect. For important work (sending an email, a final report), a human should **review and approve** first.

### Real-life analogy
A junior writes a letter, but the **manager signs off** before it is sent.

## How to add a human check

Just set **`human_input=True`** on a task. CrewAI will **pause** and ask you to review the result before finishing.

```python
import os
# Turn OFF CrewAI's usage tracking (telemetry).
# Without this, CrewAI tries to reach telemetry.crewai.com and prints a
# scary red "ERROR ... Failed to resolve" message when it cannot connect.
# It is harmless, but this keeps the output clean.
os.environ["CREWAI_TELEMETRY_OPT_OUT"] = "true"
os.environ["OTEL_SDK_DISABLED"] = "true"

from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, Task, Crew, LLM

llm = LLM(model="gpt-4o-mini")

writer = Agent(
    role="Writer",
    goal="Write a short message",
    backstory="You write friendly messages.",
    llm=llm,
)

task = Task(
    description="Write a one-line thank-you message to a customer.",
    expected_output="A one-line thank-you message.",
    agent=writer,
    human_input=True,    # <-- pause and let a human review
)

crew = Crew(agents=[writer], tasks=[task])

# In a Jupyter notebook, use kickoff_async() with "await".
result = await crew.kickoff_async()
print(result)
```

## What you will see

After the agent writes the message, CrewAI **stops and asks you** in the terminal:
- You can **accept** it, or
- **Type feedback** ("make it warmer") and the agent will try again.

This way, nothing is finalized without your okay.

> Tip: because it waits for your typing, run this in a **terminal** (`python file.py`) rather than letting a notebook run unattended.

## Key points to remember

- **Human-in-the-loop** = a person checks/approves the agent's work.
- Turn it on with **`human_input=True`** on a task.
- CrewAI **pauses** and asks you to accept or give feedback.
- Great for important or risky outputs (emails, final reports).
- Keeps a human in control of the final result.
