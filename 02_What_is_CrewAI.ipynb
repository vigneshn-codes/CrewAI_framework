# 2. What is CrewAI?

**CrewAI** is a Python framework for building a **team of AI agents** that work together to finish a job.

The name says it all: a **crew** (team) of AI workers, each with a role, working as one.

## Real-life analogy

CrewAI is like running a **small company**:

- Each **employee** = an Agent (has a job title and a goal).
- Each **assignment** = a Task.
- The **whole team** = the Crew.
- The **way they work** (one after another, or with a manager) = the Process.

You are the **boss** — you set the goals, and the crew does the work.

## Why people use CrewAI

| Without CrewAI | With CrewAI |
|----------------|-------------|
| Write lots of glue code by hand | Ready-made Agent, Task, Crew pieces |
| Hard to make agents cooperate | Built-in teamwork |
| Manual step ordering | Choose Sequential or Hierarchical |
| No memory between steps | Built-in memory option |

## The 4 main building blocks

| Block | What it is |
|-------|------------|
| **Agent** | A worker with a role, goal, and an AI brain |
| **Task** | A single job given to an agent |
| **Crew** | The team of agents + their tasks |
| **Process** | How the work runs (in order, or managed) |

We will learn each one in its own notebook.

## A tiny taste (full code comes later)

Don't worry about the details yet — just see how short it is:
```python
from crewai import Agent, Task, Crew

writer = Agent(role="Writer", goal="Write a short poem", backstory="You love poetry.")
task = Task(description="Write a 2-line poem about the sea.", expected_output="A 2-line poem.", agent=writer)
crew = Crew(agents=[writer], tasks=[task])

result = crew.kickoff()
print(result)
```

## Key points to remember

- **CrewAI** builds a **team of AI agents** that work together.
- It is like running a small company: agents are employees, tasks are assignments.
- The 4 building blocks are **Agent, Task, Crew, Process**.
- It saves you from writing lots of glue code by hand.
- You start the whole team with one line: `crew.kickoff()`.
