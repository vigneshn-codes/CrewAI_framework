# 3. CrewAI Architecture & Core Concepts

**Architecture** just means: *how the pieces fit together.*

In CrewAI, a few simple pieces connect to form a working team.

## How the pieces connect

```
            CREW (the whole team)
              /            \
         AGENTS           TASKS
      (the workers)    (the jobs to do)
           |                |
        LLM + TOOLS     given to an agent
              \            /
               PROCESS
        (the order they run in)
```

The **Crew** holds everything. **Agents** do the work, **Tasks** describe the work, and the **Process** decides the order.

## The core concepts in one table

| Concept | Simple meaning | Real-life match |
|---------|----------------|-----------------|
| **Agent** | A worker with a role + goal | An employee |
| **Task** | One job to finish | A work assignment |
| **Tool** | An extra power for an agent | A calculator or web browser |
| **Crew** | The full team | The whole company |
| **Process** | The order of work | The office workflow |
| **LLM** | The thinking brain | The employee's knowledge |
| **Memory** | Remembering past steps | Taking notes |

## The flow of a CrewAI app

1. You create one or more **Agents**.
2. You give each agent a **Task**.
3. You put them in a **Crew** and pick a **Process**.
4. You call **`crew.kickoff()`**.
5. The agents work and return a **final result**.

Every CrewAI program follows these same 5 steps.

## Key points to remember

- **Architecture** = how the pieces fit together.
- The **Crew** holds the **Agents**, **Tasks**, and a **Process**.
- Agents use an **LLM** (brain) and optional **Tools** (powers).
- Every app follows 5 steps: create agents → give tasks → build crew → kickoff → get result.
- Learn each concept one notebook at a time, then combine them.
