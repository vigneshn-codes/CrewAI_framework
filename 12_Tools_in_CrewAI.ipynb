# 12. Tools in CrewAI

A **Tool** gives an agent an **extra power** — something the AI brain alone cannot do.

Examples: search the web, read a file, do math, open a website.

## Real-life analogy

An employee is smart, but to do real work they need tools:
- a **calculator** for math,
- a **web browser** for research,
- a **printer** for documents.

Agents are the same — tools let them act on the real world.

## Ready-made tools (from crewai-tools)

Install once: `pip install crewai-tools`

| Tool | What it does |
|------|--------------|
| `SerperDevTool` | Search the web |
| `FileReadTool` | Read a text file |
| `ScrapeWebsiteTool` | Read a web page |
| `CSVSearchTool` | Search inside a CSV |

We will use some of these in the next notebooks.

## How to give a tool to an agent

You create the tool, then pass it in `tools=[...]`.

```python
from dotenv import load_dotenv
load_dotenv()

from crewai import Agent, LLM

from crewai_tools import FileReadTool

llm = LLM(model="gpt-4o-mini")

# 1. Create the tool
read_tool = FileReadTool()

# 2. Give it to an agent
reader = Agent(
    role="File Reader",
    goal="Read files and answer about them",
    backstory="You carefully read files.",
    tools=[read_tool],     # <-- the agent now has this power
    llm=llm,
)
print("Agent now has a file-reading tool!")
```

## How the agent uses a tool

You don't call the tool yourself. The agent **decides on its own** when to use it while doing a task. That is the magic of agents — they pick the right tool for the job.

## Key points to remember

- A **Tool** gives an agent an extra power (web search, file reading, etc.).
- `crewai-tools` has many ready-made tools.
- Give tools to an agent with `tools=[tool1, tool2]`.
- The agent **decides by itself** when to use a tool.
- You can also build your **own** tools (next notebook).
